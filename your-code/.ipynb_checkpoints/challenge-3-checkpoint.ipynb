#### Now in addition to the summary statistics, we also want to visualize the distribution of `Age`. Choose the appropriate plot to sow below.

```python
#your code here
def ranger_f(x):
    if x>0 and x<=10:
        y="0-10"
    elif x>10 and x<=20:
        y="11-20"
    elif x>20 and x<=30:
        y="21-30"
    elif x>30 and x<=40:
        y="31-40"
    elif x>40 and x<=50:
        y="41-50"
    elif x>50 and x<=60:
        y="51-60"   
    elif x>60 and x<=70:
        y="61-70" 
    elif x>70 and x<=80:
        y="71-80" 
    elif x>80:
        y="80-100"
    return y

titanic["Range_Age"]=titanic["Age"].apply(ranger_f)
titanic.head()

titanic_range=titanic.pivot_table(index=["Range_Age"],aggfunc="count")
titanic_range.head(10)

plt.bar(titanic_range.index,titanic_range["Age"]);
```
